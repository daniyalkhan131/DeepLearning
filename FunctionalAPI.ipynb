```python
from keras.models import Model
```

```python
from keras.layers import *
```

```python
x=Input(shape=(3,)) #becaise we have 3 columns in table
```

```python
hidden1=Dense(128,activation='relu')(x) #we have to tell that hidden1 is getting input from which layer so (x)

hidden2=Dense(64,activation='relu')(hidden1)
```

```python
#now we do branching
output1=Dense(1,activation='linear')(hidden2) #because we are doing regression

output2=Dense(1,activation='sigmoid')(hidden2) #because we are doing classification here
```

```python
model=Model(inputs=x,outputs=[output1,output2])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model,show_shapes=True)
```

```python
from keras.layers import *
from keras.models import Model
```

```python
# define two sets of inputs
inputA = Input(shape=(32,))
inputB = Input(shape=(128,))
```

```python
# the first branch operates on the first input
x = Dense(8, activation="relu")(inputA)
x1 = Dense(4, activation="relu")(x)
```

```python
# the second branch opreates on the second input
y = Dense(64, activation="relu")(inputB)
y1 = Dense(32, activation="relu")(y)
y2 = Dense(4, activation="relu")(y1)
```

```python
# combine the output of the two branches
combined = concatenate([x1, y2])
```

```python
# apply a FC layer and then a regression prediction on the
# combined outputs
z = Dense(2, activation="relu")(combined)
z1 = Dense(1, activation="linear")(z)
```

```python
# our model will accept the inputs of the two branches and
# then output a single value
model = Model(inputs=[inputA, inputB], outputs=z1)
```

```python
from keras.utils import plot_model
plot_model(model)
```
